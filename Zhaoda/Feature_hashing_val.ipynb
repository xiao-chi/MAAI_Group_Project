```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.utils import resample
from sklearn.feature_extraction import FeatureHasher
```

```python
val_df = pd.read_csv('/content/gdrive/My Drive/we_data/validation.csv')
val_df.head()
```

```python
# list(train_df) # need to remove 'Unnamed: 0', 'urlid', 'bidid','userid'
```

```python
fh = FeatureHasher(n_features=5, input_type='string')
sp = fh.fit_transform(val_df['useragent'])
sp = pd.DataFrame(sp.toarray(), columns=['useragent1', 'useragent2', 'useragent3', 'useragent4', 'useragent5'])
val_df = pd.concat([val_df, sp], axis=1)

fh = FeatureHasher(n_features=10, input_type='string')
sp = fh.fit_transform(val_df['IP'])
sp = pd.DataFrame(sp.toarray(), columns=['IP1', 'IP2', 'IP3', 'IP4', 'IP5', 'IP6', 'IP7', 'IP8','IP9','IP10'])
val_df = pd.concat([val_df, sp], axis=1)

fh = FeatureHasher(n_features=10, input_type='string')
sp = fh.fit_transform(val_df['domain'].fillna('Unknown'))
sp = pd.DataFrame(sp.toarray(), columns=['domain1', 'domain2', 'domain3', 'domain4', 'domain5', 'domain6', 'domain7', 'domain8','domain9','domain10'])
val_df = pd.concat([val_df, sp], axis=1)

fh = FeatureHasher(n_features=10, input_type='string')
sp = fh.fit_transform(val_df['url'].fillna('Unknown'))
sp = pd.DataFrame(sp.toarray(), columns=['url1', 'url2', 'url3', 'url4', 'url5', 'url6', 'url7', 'url8','url9','url10'])
val_df = pd.concat([val_df, sp], axis=1)

fh = FeatureHasher(n_features=10, input_type='string')
sp = fh.fit_transform(val_df['slotid'])
sp = pd.DataFrame(sp.toarray(), columns=['slotid1', 'slotid2', 'slotid3', 'slotid4', 'slotid5', 'slotid6', 'slotid7', 'slotid8','slotid9','slotid10'])
val_df = pd.concat([val_df, sp], axis=1)

fh = FeatureHasher(n_features=5, input_type='string')
sp = fh.fit_transform(val_df['creative'])
sp = pd.DataFrame(sp.toarray(), columns=['creative1', 'creative2', 'creative3', 'creative4', 'creative5'])
val_df = pd.concat([val_df, sp], axis=1)

fh = FeatureHasher(n_features=3, input_type='string')
sp = fh.fit_transform(val_df['keypage'].fillna('Unknown'))
sp = pd.DataFrame(sp.toarray(), columns=['keypage1', 'keypage2', 'keypage3'])
val_df = pd.concat([val_df, sp], axis=1)
```

```python
feature_hash = ['useragent','IP','domain','url','slotid','creative','keypage']
val_df = val_df.drop(feature_hash, axis = 1)
```

```python
cat_vars = ['weekday','region','adexchange','slotvisibility','slotformat','advertiser']

for i in cat_vars:
  one_hot = pd.get_dummies(val_df[i],prefix=i)
  val_df = val_df.drop(i,axis = 1)
  val_df = val_df.join(one_hot)

val_df.head()
```
